# Personal Plotting Style

Below is an example of how I setup all of my plots for publication. Gives decent looking plots for astronomical journals.

```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

plot_style = {'axes.axisbelow': True,
             'axes.edgecolor': '0',
             'axes.facecolor': 'white',
             'axes.grid': False,
             'axes.labelcolor': '.15',
             'axes.spines.bottom': True,
             'axes.spines.left': True,
             'axes.spines.right': True,
             'axes.spines.top': True,
             'figure.facecolor': 'white',
             'font.family': ['serif'],
             'font.sans-serif': ['Times New Roman'],
             'grid.color': '.8',
             'grid.linestyle': '-',
             'image.cmap': 'viridis',
             'lines.solid_capstyle': 'round',
             'patch.edgecolor': 'w',
             'patch.force_edgecolor': True,
             'text.color': '.15',
             'xtick.bottom': True,
             'xtick.color': '.15',
             'xtick.direction': 'in',
             'xtick.top': True,
             'ytick.color': '.15',
             'ytick.direction': 'in',
             'ytick.left': True,
             'ytick.right': True}

axes_fontsize = 16
figsize = (6.4, 5.5)
dpi = 110

sb.set_style(plot_style)
colors = sb.color_palette('plasma', 4)
```

## Example

Here is an example with 4 sin functions using my selected 4 colours.

```python
x = np.linspace(0, 2 * np.pi, 100)

plt.figure(figsize = figsize, dpi = dpi, facecolor = 'w')
plt.xlim([0, 2 * np.pi])
plt.ylim([-1, 1])
plt.xlabel('x', fontsize = axes_fontsize)
plt.ylabel('y', fontsize = axes_fontsize)

for i in range(0, 3):
    plt.plot(x, np.sin((i + 1) * x), c = colors[i])

plt.show()
plt.close()
```
